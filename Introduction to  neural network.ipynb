# 1.Activation functions

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0,x)
```

# 2.Sample data

```python
x1=3
x2=2
y_actual=17
```

# 3.Feed Forward

```python
np.random.seed(42)
w1=np.random.uniform(1,10)
w2=np.random.uniform(1,10)

learning_rate=0.01
```

# 4.Error Calculation

```python
def calculate_error(y_actual, y_pred):
    return (y_actual - y_pred) ** 2
```

# 5.Back Propogation

```python
def update_weights(x1, x2, y_actual, y_pred, w1, w2, learning_rate):
    # Gradient of the error with respect to the predicted output
    grad_y_pred = -2 * (y_actual - y_pred)
    
    # Gradients of the error with respect to weights
    grad_w1 = grad_y_pred * x1
    grad_w2 = grad_y_pred * x2
    
    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
    
    return w1, w2
```

# 6.Training

```python
# Number of training iterations (epochs)
epochs = 20

# To store error progression
error_progression = []
ypred_progression = []

for epoch in range(epochs):
    # Feed forward
    y_pred = x1 * w1 + x2 * w2
    
    # Calculate error
    error = calculate_error(y_actual, y_pred)
    
    # Record error and prediction for visualization
    error_progression.append(error)
    ypred_progression.append(y_pred)
    
    # Back propagation and weight update
    w1, w2 = update_weights(x1, x2, y_actual, y_pred, w1, w2, learning_rate)
    
    print(f"Epoch {epoch+1}: y_pred={y_pred}, error={error}, w1={w1}, w2={w2}")
```

# 7.Visualization

```python
import matplotlib.pyplot as plt

# Plot the progression of error
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(epochs), error_progression, label='Error')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error Progression Across Epochs')
plt.legend()

# Plot the progression of y_pred
plt.subplot(1, 2, 2)
plt.plot(range(epochs), ypred_progression, label='Predicted y', color='orange')
plt.axhline(y=y_actual, color='red', linestyle='--', label='Actual y')
plt.xlabel('Epochs')
plt.ylabel('y_pred')
plt.title('Prediction Progression Across Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```
